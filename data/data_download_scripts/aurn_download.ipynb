```python
from pyaurn import importAURN, importMeta, timeAverage

# Download metadata of site IDs, names, locations etc
metadata = importMeta(source="aurn")

print(metadata)
```

```python
urls=[]
for year in range(2022,2026):
    for site_id in metadata['site_id']:
        url = f"https://uk-air.defra.gov.uk/openair/R_data/{site_id}_{year}.RData"
        urls.append(url)
urls
```

```python
with open('aurn_urls.txt','w') as f:
    f.write('\n'.join(urls))
```
