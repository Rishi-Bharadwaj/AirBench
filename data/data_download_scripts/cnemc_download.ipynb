```python
import pandas as pd

dates=pd.date_range(start="2014-05-13", end="2025-12-31", freq="D")
dates= dates.strftime("%Y%m%d").to_list()
```

```python
url_list=[]
for date in dates:
    url=f"https://quotsoft.net/air/data/china_sites_{date}.csv"
    url_list.append(url)
url_list

with open("cnemc_urls.txt", 'w') as f:
    f.write("\n".join(url_list))
```
