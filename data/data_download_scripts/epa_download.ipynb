```python
urls=[]
years=range(2000,2026)
# codes=[44201,42401,42101,42602,88101,81102, "WIND","TEMP" ,"PRESS", "RH_DP"]
codes=[44201,42401,42101,42602,88101,81102]
```

```python
for year in years:
    for code in codes:
        urls.append(f"https://aqs.epa.gov/aqsweb/airdata/hourly_{code}_{year}.zip")
urls
```

```python
with open("epa_urls.txt", "w") as f:
    f.write("\n".join(urls))
```
